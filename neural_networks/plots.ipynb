## Feedforward Neural Network Plot
With bias nodes, two inputs, two hidden, one output.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes with positions
G.add_node('B1', pos=(0, 3), label='Bias')
G.add_node('I1', pos=(0, 2), label=r'$x_1$')
G.add_node('I2', pos=(0, 1), label=r'$x_2$')

G.add_node('B2', pos=(1, 3), label='Bias')
G.add_node('H1', pos=(1, 2), label=r'$h_1$')
G.add_node('H2', pos=(1, 1), label=r'$h_2$')

G.add_node('O', pos=(2, 2), label='Output')

# Alternative labels
# G.add_node('B1', pos=(0, 3), label='Bias')
# G.add_node('I1', pos=(0, 2), label='Input 1')
# G.add_node('I2', pos=(0, 1), label='Input 2')

# G.add_node('B2', pos=(1, 3), label='Bias')
# G.add_node('H1', pos=(1, 2), label='Hidden 1')
# G.add_node('H2', pos=(1, 1), label='Hidden 2')

# G.add_node('O', pos=(2, 2), label='Output')

# Add edges with weights as labels
edges = [
    ('B1', 'H1', r'$b_1$'), ('B1', 'H2', r'$b_2$'),
    ('I1', 'H1', r'$w_1$'), ('I1', 'H2', r'$w_2$'),
    ('I2', 'H1', r'$w_3$'), ('I2', 'H2', r'$w_4$'),
    ('H1', 'O', r'$w_5$'), ('H2', 'O', r'$w_6$'),
    ('B2', 'O', r'$b_3$')
]

for edge in edges:
    G.add_edge(edge[0], edge[1], weight=edge[2])

# Get node positions
pos = nx.get_node_attributes(G, 'pos')

# Draw the nodes
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', labels=nx.get_node_attributes(G, 'label'))

# Draw the edges
nx.draw_networkx_edges(G, pos)

# Draw the edges with weights, using a custom bbox to adjust positions
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
bbox_props = dict(boxstyle="round,pad=0.3", edgecolor='none', facecolor='white', alpha=1.0)

# Define positions for the edge labels
edge_label_pos = {
    ('B1', 'H1'): (0.5, 2.5),
    ('B1', 'H2'): (0.3, 2.4),

    ('I1', 'H1'): (0.3, 1.98),
    ('I1', 'H2'): (0.25, 1.74),

    ('I2', 'H1'): (0.25, 1.23),
    ('I2', 'H2'): (0.3, 0.98),

    ('B2', 'O'): (1.5, 2.5),
    ('H1', 'O'): (1.5, 1.98),
    ('H2', 'O'): (1.5, 1.49)
}

for (u, v), label in edge_labels.items():
    x, y = edge_label_pos[(u, v)]
    plt.text(x, y, label, bbox=bbox_props, horizontalalignment='center')


# Add custom text labels for layers
plt.text(0, 3.35, 'Input Layer', fontsize=12, ha='center', va='center', weight='bold')
plt.text(1, 3.35, 'Hidden Layer', fontsize=12, ha='center', va='center', weight='bold')
plt.text(2, 3.35, 'Output Layer', fontsize=12, ha='center', va='center', weight='bold')

# Show plot
plt.show()
```

## Simple Neural Network Plot
With one bias node, two inputs, one output.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes with positions
G.add_node('B1', pos=(0, 3), label='Bias')
G.add_node('I1', pos=(0, 2), label=r'$x_1$')
G.add_node('I2', pos=(0, 1), label=r'$x_2$')

G.add_node('O', pos=(1, 1.5), label='Output')

# Add edges with weights as labels
edges = [
    ('B1', 'O', r'$b_1$'),
    ('I1', 'O', r'$w_1$'),
    ('I2', 'O', r'$w_2$'), 
]

for edge in edges:
    G.add_edge(edge[0], edge[1], weight=edge[2])

# Get node positions
pos = nx.get_node_attributes(G, 'pos')

# Draw the nodes
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', labels=nx.get_node_attributes(G, 'label'))

# Draw the edges
nx.draw_networkx_edges(G, pos)

# Draw the edges with weights, using a custom bbox to adjust positions
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
bbox_props = dict(boxstyle="round,pad=0.3", edgecolor='none', facecolor='white', alpha=1.0)

# Define positions for the edge labels
edge_label_pos = {
    ('B1', 'O'): (0.5, 2.23),
    ('I1', 'O'): (0.5, 1.73),
    ('I2', 'O'): (0.5, 1.23),
}

for (u, v), label in edge_labels.items():
    x, y = edge_label_pos[(u, v)]
    plt.text(x, y, label, bbox=bbox_props, horizontalalignment='center')

# Add custom text labels for layers
plt.text(0, 3.35, 'Input Layer', fontsize=12, ha='center', va='center', weight='bold')
plt.text(1, 3.35, 'Output Layer', fontsize=12, ha='center', va='center', weight='bold')

# Show plot
plt.show()
```
